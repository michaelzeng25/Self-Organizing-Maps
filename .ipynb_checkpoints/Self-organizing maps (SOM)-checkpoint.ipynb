```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv('Credit_Card_Applications.csv')
```

```python
df.head()
```

```python
X = df.iloc[:, :-1].values
y = df.iloc[:,-1].values
```

```python
# feature scalling to make it easier for the model to be trained

from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler(feature_range=(0,1))
```

```python
X = scaler.fit_transform(X)
```

```python
# training the SOM
```

```python
from minisom import MiniSom
```

```python
som = MiniSom(x=10, y=10, input_len=15, ) # 15 features in X
```

```python
som.random_weights_init(X)
```

```python
som.train_random(data = X, num_iteration=100)
```

```python
# visualize the results
```

```python
from pylab import bone, pcolor, colorbar, plot, show
```

```python
# initilize the figure- a empty window
bone()

# distance map will return all the mean into neuron distance in one matrix
# T = transpose
pcolor(som.distance_map().T)

# show the scale
colorbar()

# red 'o' = customers did not get approval. green 's' = opposite
marker = ['o','s']
colors = ['r','g']

for i, x in enumerate(X):
    winning_node = som.winner(x)
    
    #put the marker in the center of the winning node
    plot(winning_node[0] + 0.5,
         winning_node[1] + 0.5,
         marker[y[i]], # if y[i]=0 it will be the 'o' marker
         markeredgecolor = colors[y[i]],
        markerfacecolor='None',
        markersize=10,
        markeredgewidth=2)
show()
```

```python
# among those potential frauds, some got approval because we got both cases.
# index for those outliers are: 7 and 1, 7 and 2
```

```python
bone()
pcolor(som.distance_map().T)
colorbar()
marker = ['o','s']
colors = ['r','g']

for i, x in enumerate(X):
    winning_node = som.winner(x)
    plot(winning_node[0] + 0.5,
         winning_node[1] + 0.5,
         marker[y[i]], 
         markeredgecolor = colors[y[i]],
        markerfacecolor='None',
        markersize=10,
        markeredgewidth=2)
show()
```

```python
mappings = som.win_map(X)
frauds = np.concatenate( ( mappings[(7,1)], mappings[(7,2)]), axis=0)
frauds
```

```python
# inverse the scale
frauds = scaler.inverse_transform(frauds)
frauds
```

```python
# get the customer ids of those potential frauds
for i in range(len(frauds)):
    print(int(frauds[i][0]))
```
